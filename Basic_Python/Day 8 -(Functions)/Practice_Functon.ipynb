```python
def Quarters(User_date):
  date_year = User_date[:4]
  print(date_year)

  date_month = int(User_date[5:7])
  print(date_month)
  if date_month <= 3:
    print('2020-Q1')

  elif date_month <= 6:
    print('2020-Q2')

  elif date_month <= 9:
    print('2020-Q3')

  else:
    print('2020-Q4')
```

```python
a = '2020-07-20'
Quarters(a)
```

```python
b = '2020-12-20'
Quarters(b)
```
